```python
sd_dataset = sd.load_dataset()
```

```python
df_sample = pd.read_csv(sd_dataset.samples_csv)
df_sample
```

```python
df_curve = pd.read_csv(sd_dataset.curves_csv)
df_curve
```

```python
df = pd.merge(df_sample, df_curve, on=['sample_id', 'SID', 'DOI'], suffixes=('_sample', '_curve'))
df
```

```python
df.columns
```

```python
df_2 = df.rename(columns={
    'composition_sample': 'composition',
})
```

```python
df_2.columns
```

```python
df_te = df_2[df_2['project_names'] == '["ThermoelectricMaterials"]'].reset_index(drop=True)
df_te['project_names'].value_counts()
```

```python
import json
from tqdm import tqdm

# ユニークなdescriptorを収集するセット
unique_descriptors = set()

# すべての行をスキャンしてユニークなdescriptorを収集
for index, row in tqdm(df_te.iterrows(), total=df_te.shape[0]):
    try:
        sample_info_json = json.loads(row['sample_info'])
        for descriptor in sample_info_json.keys():
            unique_descriptors.add(descriptor.replace(" ", "_").lower())
    except json.JSONDecodeError as error:
        # "JSONファイルの読み込みに失敗しましたが、Descriptorのuniqueリストを作成するだけなので、続行
        continue
    except TypeError as error:
        # "JSONファイルの読み込みに失敗しましたが、Descriptorのuniqueリストを作成するだけなので、続行
        continue
    # except Exception as error:
    #     print(index, row['sample_info'])
    #     print(f"予期せぬエラーが発生しました: {error}")
    #     raise error
```

```python
unique_descriptors
```

```python
import re
import numpy as np
import math

# 新しい列を格納する辞書
new_columns = {f"{descriptor}_category": [None] * df_te.shape[0] for descriptor in unique_descriptors}
new_columns.update({f"{descriptor}_comment": [None] * df_te.shape[0] for descriptor in unique_descriptors})

# すべての行を処理して、新しい列にデータを追加
for index, row in tqdm(df_te.iterrows(), total=df_te.shape[0]):
    if pd.isna(row['sample_info']):
        continue
    sample_info_json = json.loads(row['sample_info'])
    for descriptor, value_object in sample_info_json.items():
        descriptor = descriptor.replace(" ", "_").lower()  # keyを変数として使用
        if "category" in value_object:
            if re.search(r"family", descriptor):
                new_columns["material_family_category"][index] = value_object["category"]
            elif re.search(r"grain", descriptor):
                new_columns["grain_size_category"][index] = value_object["category"]
            elif re.search(r"maximum_power", descriptor):
                new_columns["maximum_power_category"][index] = value_object["category"]
            elif re.search(r"thermal", descriptor):
                new_columns["thermalmeasurement_category"][index] = value_object["category"]
            else:
                new_columns[f"{descriptor}_category"][index] = value_object["category"]
        if "comment" in value_object:
            if re.search(r"family", descriptor):
                new_columns["material_family_comment"][index] = value_object["comment"]
            elif re.search(r"grain", descriptor):
                new_columns["grain_size_comment"][index] = value_object["comment"]
            elif re.search(r"maximum_power", descriptor):
                new_columns["maximum_power_comment"][index] = value_object["comment"]
            elif re.search(r"thermal", descriptor):
                new_columns["thermalmeasurement_comment"][index] = value_object["comment"]
            else:
                new_columns[f"{descriptor}_comment"][index] = value_object["comment"]
    # except Exception as error:
    #     # JSON読み込みエラーは無視して続行
    #     continue
    #     raise error

# 新しい列を一度にデータフレームに追加
df_new_columns = pd.DataFrame(new_columns)
df_te_2 = pd.concat([df_te.reset_index(drop=True), df_new_columns], axis=1)
```

```python
df_te_3 = df_te_2.dropna(axis=1, how='all')
df_te_3.shape
```

```python
# df_te_3.columns.str.contains("category")
categories = df_te_3.columns[df_te_3.columns.str.contains("category")]
categories
```

```python
import matplotlib.pyplot as plt
import os

# ファイル名を安全にする関数
def sanitize_filename(filename):
    return re.sub(r'[\/:*?"<>|]', '_', filename)


for cat in categories:
    value_counts = df_te_3[cat].value_counts()
    num_values = len(value_counts)

    # 図を設定
    plt.figure(figsize=(10, num_values * 0.3))
    value_counts.plot.barh(title=f"{cat} (Date: {sd_dataset.dataset_timestamp})")


    # カテゴリ名を安全なファイル名に変換
    safe_cat = sanitize_filename(cat)

    # 画像のファイル名を設定
    # file_path = os.path.join("./image/", f"{safe_cat}_barh.png")

    # CSVのファイル名を設定
    csv_path = os.path.join("./csv/", f"{safe_cat}_values.csv")

    # 値をCSVとして保存
    value_counts.to_csv(csv_path, header=True, encoding='utf-8-sig')

    # グラフを画像として保存
    plt.show()
    plt.close()  # グラフを閉じる
```
